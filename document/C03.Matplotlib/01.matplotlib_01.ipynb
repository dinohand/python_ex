```python
# 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

lstyle = 'dashed'             # https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
lcolor = 'red'                # https://matplotlib.org/stable/users/explain/colors/colors.html
lwidth = 2.5
marker = '>' #  o, *, p, >   # https://matplotlib.org/stable/api/markers_api.html
markersize = 10

x = [10, 20, 30, 40, 50]
y = [1, 2, 3, 4, 5]
plt.plot(x, y,  linestyle=lstyle, color=lcolor, linewidth=lwidth, marker=marker, markersize=markersize)
plt.title('Line Plot Example')
plt.xlabel('X-axis Label') 
```

```python
lstyle = 'dashed'
lcolor = "#ff00ff" ## ff00ff : magenta , #ffff00  ## yellow

x = [10, 20, 30, 40, 50]
y = [1, 2, 3, 4, 5]
# plt.plot(x, y, marker='*', linestyle=lstyle, color=lcolor)
plt.plot(x, y, color = lcolor, label ='line')   ## 
plt.legend( loc=10 ) ## loc = [0,1,2,3,5,6,7,8,9, 10]  ## 0: best, 1: upper right, 2: upper left, 3: lower left, 4: lower right, 5: right, 6: center left, 7: center right, 8: lower center, 9: upper center, 10 : center



plt.title('Line Plot Example')
plt.xlabel('X-axis Label') 


plt.savefig('line_plot_example.png', dpi=300, bbox_inches='tight')
```
