```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [1, 2, 1.5, 4, 5]
y1 = [5, 4, 3, 2, 4]

plt.plot(x,y, label='1번 데이터')
plt.plot(x,y1, label='2번 데이터')

plt.rcParams['font.family'] = 'Malgun Gothic'

# title, labels
plt.title('나는 제목', fontdict={'size':20, 'weight':'bold'} , color='blue', loc='center')
plt.ylabel('나는 Y 라벨')
plt.xlabel('나는 X 라벨')
# plt.legend( title= '범례', loc=7 ) 
plt.legend( title= '범례', loc='upper center') 


plt.show()
```
