```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(4) # 0,1,2,3
years = ['2018', '2019', '2020','2021']
values = [100, 400, 900, 1200]

plt.title('Bar Chart')  
plt.bar(x, values, color=['red','yellow','blue', 'green'],  width=0.5 )
plt.xticks(ticks=x, years)  # x축  눈금 설정 함수

plt.show()
```
