```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2018', '2019', '2020','2021']   # x 좌표에 표시될 값
values = [100, 400, 900, 1200]            # y 좌표에 표시될 값
bottoms = [100, 200.2, 50.50, 400]
colors=['red','yellow','blue', 'green']
unicolor = "green"


plt.title('Bar Chart')  
plt.bar(years,  values, color = colors, bottom=bottoms) 
# plt.barh(years,  values, color = colors, left=bottoms) 

plt.xlabel('Years')
plt.ylabel('Values')

plt.show()
```
