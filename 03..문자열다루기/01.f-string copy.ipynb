```python
# f-string
# 문자열 안에 변수값을 대입하는 방법
# f"문자열 {변수1} {변수2}"와 같이 변수값사용할 곳에 {변수명}과 같이 사용한다
# {}에 들어갈 변수는 미리 정의되어 있어야 한다
# 만약 문자열에 '{' 또는 '}'을 출력하고 싶다면, '{{' 또는 '}}'으로 표현하면 된다

who  = "갑돌이"
whom = "갑순이"
loc = ['서울', '부산', '제주']
       
msg  = f"'{who}'가 {loc[0]}에서 {{{whom}}}에게 데이트를 신청했다"

print(msg)

# 중요
# f-string안의 값으로 표현될 수 있는 코드가 들어갈 수 있다

that = True

print(f"1 + 1 = { 1 + 1 } 입니다")

print(f"f-string에서 if-else도 사용할 수 있습니다 => {'성공' if that else '실패' }")

## 숫자 출력 형식
# 3 자리마다, 지정된 구분기호를 삽입한다다 
number = 1234567.89
print(f"{number:,}")

# 숫자 출력 자릿수 지정 => {값 : 자릿수}
# 3 : = 3자릿수 숫자로 출력한다. 단 3자릿수가 안되면 공백(' ')츨력한다
# 03 : 빈자리를 '0'으로 채운다다
# 지정된 자릿수를 가진 숫자는 trim하지 않고 그대로 출력한다다
nl = [1,1.1, 22, 22.3, 333, 333.3, 4444, 55555]

for item in nl:
    print(f"{item:4}")
    
for item in nl:
    print(f"{item:04}")
    
# 실수의 소숫점 지정
# {값 : .자릿수f}
fn = -123.456789

print(f"{fn:.2f}" , f"{-123.456789:3f}" )


# 지수 표기법 (Exponential notaion
a = 2000.0

print(f"{a:.2}")


# 날짜 출력 형식 지정정
import datetime

print(f"현재 시간은 '{ datetime.datetime.now() }' 입니다")
print(f"현재 시간은 '{ datetime.datetime.now():%Y년 %m월 %d일}' 입니다")

##### 참조 : https://m.blog.naver.com/youji4ever/222429288222
```
